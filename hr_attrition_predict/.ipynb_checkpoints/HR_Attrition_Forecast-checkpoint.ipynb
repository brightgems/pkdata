## feature Generiation
- Age -> AgeGroup: [young <30,middle <40, adult]
- DistanceFromHome -> HomeDistanceGroup: [nearby <10, faraway]


```python
# agegroup
def to_agegroup(x):
    if x.Age<=35:
        return 'young'
    elif x.Age<=45:
        return 'middle age'
    else:
        
        return 'adult'
        
HR_emp_att['AgeGroup']= pd.Categorical(HR_emp_att.apply(to_agegroup,axis=1))

# HomeDistanceGroup
HR_emp_att['HomeDistanceGroup']= pd.Categorical(HR_emp_att.apply(lambda x: 'nearby' if x.DistanceFromHome<=10 else 'faraway',axis=1))
HR_emp_att['Satisfaction'] = HR_emp_att.apply(lambda x: x.EnvironmentSatisfaction+x.JobInvolvement+x.JobSatisfaction+x.RelationshipSatisfaction+x.WorkLifeBalance,axis=1)/5
HR_emp_att['Log_DistanceFromHome'] = HR_emp_att.DistanceFromHome.apply(np.log)
HR_emp_att['Log_MonthlyIncome'] = HR_emp_att.MonthlyIncome.apply(np.log)
HR_emp_att['Log_SalaryHike'] = np.log(HR_emp_att.PercentSalaryHike * HR_emp_att.MonthlyIncome)
HR_emp_att.groupby(['JobRole','Attrition'])['YearsAtCompany'].agg(['mean','count'])
```

## Variable Transformations

```python
X_train = HR_emp_att.drop(['Attrition'], axis=1)
y_train = HR_emp_att.Attrition

from sklearn.feature_extraction import DictVectorizer

def one_hot_dataframe(data, cols, replace=True):
    vec = DictVectorizer()
    mkdict = lambda row: dict((col, row[col]) for col in cols)
    
    vecData = pd.DataFrame(vec.fit_transform(data[cols].apply(mkdict, axis=1)).toarray())
    vecData.columns = vec.get_feature_names()
    vecData.index = data.index
    
    if replace is True:
        data = data.drop(cols, axis=1)
    data = data.join(vecData)
    return (data, vecData, vec)

def factorize_dataframe(data,cols):
    for col in cols:
        data[col] = pd.factorize(data[col])[0] # factorize return 2 elements array, values and category index
    return data
        

cat_vars = X_train.dtypes.index[X_train.dtypes=='category']
X_train = factorize_dataframe(X_train, cat_vars)
# X_train,_,_ = one_hot_dataframe(X_train,cat_vars,replace=True)
```

## detect outlier
Isolation Forest is an anomaly detection algorithm based on trees. The algorithm builds a number of random trees and the rationale is that if a sample is isolated it should alone in a leaf after very few random splits. Isolation Forest builds a score of abnormality based the depth of the tree at which samples end up.

```python
from sklearn.ensemble import IsolationForest

iforest = IsolationForest(n_estimators=300, contamination=0.05)
iforest = iforest.fit(X_train)
iforest_pred = iforest.predict(X_train) 
X_outliers = X_train[iforest_pred== -1]

X_cleaned = X_train[iforest_pred==1]
y_cleaned = y_train[iforest_pred==1]
print(X_outliers.shape)
@interact(x=X_outliers.columns)
def compare_dist(x):
    plt.figure(figsize=(18,6))
    sns.distplot(X_outliers[x]);
    plt.show();
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split   # cross_validation
from sklearn.grid_search import RandomizedSearchCV,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, make_scorer
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.feature_selection import RFECV,SelectKBest
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB  
from sklearn.metrics import roc_curve,confusion_matrix
from sklearn.metrics import roc_auc_score  
from sklearn.metrics import classification_report 
from sklearn.pipeline import make_pipeline, Pipeline, FeatureUnion
from sklearn.cross_validation import StratifiedKFold
from sklearn.decomposition import PCA
from tempfile import mkdtemp
from shutil import rmtree
from imblearn.over_sampling import RandomOverSampler, SMOTE
from sklearn.ensemble import  AdaBoostClassifier
from xgboostwrapper import XGBoostClassifier

# parameter_space = {  
#     "randomforestclassifier__n_estimators": [10, 15, 20],  
#     "randomforestclassifier__criterion": ["gini", "entropy"],  
#     "randomforestclassifier__min_samples_leaf": [2, 4, 6], 
#     "randomforestclassifier__max_depth" : [2,3,4],
#     "randomforestclassifier__min_samples_split" : [5,10,15]
# }  
#  Create a logistic regression estimator 
cachedir = mkdtemp()

ros = SMOTE(random_state=0)
X_resampled, y_resampled = ros.fit_sample(X_train, y_train)
# X_resampled, y_resampled = ros.fit_sample(X_cleaned, y_cleaned)

lrc_model = LogisticRegression(penalty='l1',C=0.05)
estimator = XGBoostClassifier(num_class=2, nthread=4, silent=1,num_boost_round=250, eta=0.05, max_depth=4)

xgb_parameters = {
#         'clf__num_boost_round': [100, 250, 500],
        'clf__eta': [0.01, 0.1, 0.5,1],
         'clf__max_depth': [4, 6],
#         'clf__subsample': [0.9, 1.0],
#         'clf__colsample_bytree': [0.9, 1.0],
    
    }    

# Use RFECV to pick best features, using Stratified Kfold
rfecv =   RFECV(estimator=lrc_model,step=1,cv=5, scoring='f1')


# This dataset is way too high-dimensional. Better do PCA:
pca = PCA(n_components=2)

# Maybe some original features where good, too?
selection = SelectKBest(k=1)

# Build estimator from PCA and Univariate selection:

combined_features = FeatureUnion([('rfecv',rfecv),("pca", pca)]) #,("univ_select", selection),("pca", pca)

pipe = Pipeline([('scale',StandardScaler()),("features", combined_features),('clf',estimator)])
# Define a range of hyper parameters for grid search
C_range = list(np.linspace(.1,10,10))
penalty_options = ['l1', 'l2']
parameter_space = dict(#features__pca__n_components=[2],
#                   features__univ_select__k=[5,7,10,13],
                  clf__learning_rate=(1, 0.1, 0.01)
                    )
parameter_space.update(xgb_parameters)
                  

grid_search = GridSearchCV(pipe , parameter_space, n_jobs=2,verbose=1,scoring='f1')
grid_search.fit(X_resampled,y_resampled)
print("Best score: %0.3f" % grid_search.best_score_)

# Clear the cache directory when you don't need it anymore
rmtree(cachedir)
```

```python
def prepare_dataset(HR_emp_att):
    # 删除Over18/ StandardHours
    if features_has_same_val.index.values[0] in HR_emp_att.columns:
        HR_emp_att = HR_emp_att.drop(features_has_same_val.index, axis=1)
    # 由于缺少员工编号的上下文背景，删除EmployeeNumber
    if 'EmployeeNumber' in HR_emp_att.columns:
        HR_emp_att = HR_emp_att.drop('EmployeeNumber', axis=1)
    HR_emp_att.Education = pd.Categorical(HR_emp_att.Education)
    HR_emp_att.EnvironmentSatisfaction = pd.Categorical(HR_emp_att.EnvironmentSatisfaction)
    HR_emp_att.JobLevel = pd.Categorical(HR_emp_att.JobLevel)
    HR_emp_att.JobSatisfaction = pd.Categorical(HR_emp_att.JobSatisfaction)
    HR_emp_att.PerformanceRating = pd.Categorical(HR_emp_att.PerformanceRating)
    HR_emp_att.RelationshipSatisfaction = pd.Categorical(HR_emp_att.RelationshipSatisfaction)
    HR_emp_att.StockOptionLevel = pd.Categorical(HR_emp_att.StockOptionLevel)
    HR_emp_att.WorkLifeBalance = pd.Categorical(HR_emp_att.WorkLifeBalance)
    
    for obj_var in HR_emp_att.select_dtypes(['object']).columns.values:
        HR_emp_att[obj_var] = pd.Categorical(HR_emp_att[obj_var])
        
    HR_emp_att['AgeGroup']= pd.Categorical(HR_emp_att.apply(to_agegroup,axis=1))
    
    # HomeDistanceGroup
    HR_emp_att['HomeDistanceGroup']= pd.Categorical(HR_emp_att.apply(lambda x: 'nearby' if x.DistanceFromHome<=10 else 'faraway',axis=1))
    HR_emp_att['Satisfaction'] = HR_emp_att.apply(lambda x: x.EnvironmentSatisfaction+x.JobInvolvement+x.JobSatisfaction+x.RelationshipSatisfaction+x.WorkLifeBalance,axis=1)/5
    HR_emp_att['Log_DistanceFromHome'] = HR_emp_att.DistanceFromHome.apply(np.log)
    HR_emp_att['Log_MonthlyIncome'] = HR_emp_att.MonthlyIncome.apply(np.log)
    HR_emp_att['Log_SalaryHike'] = np.log(HR_emp_att.PercentSalaryHike * HR_emp_att.MonthlyIncome)
    # feature selection
#     HR_emp_att = HR_emp_att[sel_features]
    cat_vars = HR_emp_att.dtypes.index[HR_emp_att.dtypes=='category']
    HR_emp_att = factorize_dataframe(HR_emp_att, cat_vars)
    
    return HR_emp_att
```
