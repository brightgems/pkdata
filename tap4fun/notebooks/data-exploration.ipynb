整理resource/army/accelation

reduce% = (add - reduce)/add

```python
 df_trans.loc[:,df_variable[df_variable.cat=='army'].column.values].describe()
```

```python
from toolz.curried import pipe, map, filter, get

def merge_add_reduce(cat):
    global df_variable, df_trans, df_train
    vars = df_variable[df_variable.cat==cat].column
    vars_add = pipe(vars, filter(lambda x: x.find('add_')>0),list)
    vars_rdc = pipe(vars, filter(lambda x: x.find('reduce_')>0),list)
    df_trans[cat+'_add']= df_train[vars_add].sum(axis=1)
    df_trans[cat+'_reduce'] = df_train[vars_rdc].sum(axis=1)
    subcats = np.unique([str.join('_',a.split('_')[:-2]) for a in vars])
    print(subcats)
    for each in subcats:
        try:
            df_trans[each+'_reduce%']= df_trans[each+'_reduce_value']/(df_trans[each+'_add_value']+1)
            df_trans[each+'_net']= df_trans[each+'_add_value'] - df_trans[each+'_reduce_value']
        except:
            pass
    
#     try:
#         df_trans = df_trans.drop(vars, axis=1)
#     except:
#         pass
    
cat_set = df_variable[df_variable.column.str.find('reduce')>0].cat.unique()

for cat in cat_set:
    merge_add_reduce(cat)
```

```python
# factorize
def factorize_dataframe(data,cols):
    for col in cols:
        if data[col].dropna().drop_duplicates().shape[0]>2:
            df_dummy = pd.get_dummies(data[col],prefix=col) # factorize return 2 elements array, values and category index
            data = pd.concat([data,df_dummy],axis=1)
            data=data.drop(col,axis=1)
        else:
            data[col] = pd.factorize(data[col])[0]
    return data

cat_vars = X.select_dtypes(['category']).columns
X = factorize_dataframe(X, cat_vars)
```
